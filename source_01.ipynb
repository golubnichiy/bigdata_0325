### Фильтры в pandas

Далее будут **примеры** использования фильтров в `pandas`. 

```python
print(dogs['Height (cm)'] > 50)
```

```python
dogs[dogs['Height (cm)'] > 50]
```

```python
dogs[dogs['Breed'] == 'Лабрадор']
```

```python
# str_1 = 'это очень "важный" текст'
str_1 = "это очень \"важный\" текст"
str_2 = "I don't speak English"
print(str_2)
```

```python
dogs[dogs['Date of Birth'] > '2017-11-20'] # 2025-03-31 13:21:34
```

```python
is_small = dogs['Height (cm)'] < 50
is_black = dogs['Color'] == 'Черный'
dogs[is_small | is_black]
```

```python
is_black_or_white = dogs['Color'].isin(['Черный', 'Белый'])
dogs[is_black_or_white]
```

```python
dogs['Height (m)'] = dogs['Height (cm)'] / 100
dogs.head()
```

```python
dogs.shape
```

```python
dogs['BMI'] = dogs['Weight (kg)'] / dogs['Height (m)'] ** 2
```

```python
dogs.head()
```

```python
dogs.describe()
```

```python
dogs['Height (cm)'].mean()
```

```python
def pct30(column):
    return column.quantile(0.3)

def pct70(column):
    return column.quantile(0.7)

dogs[['Weight (kg)', 'Height (m)', 'BMI']].agg([pct30, pct70])
```
