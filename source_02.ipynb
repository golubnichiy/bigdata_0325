### Подключаем библиотеки

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.width', 200)
```

### Загружаем данные

```python
data = pd.read_csv('HWS.csv')
print(data.head())
print(data.shape)
```

### Гистограмма

```python
data['Height'].hist()
plt.show()
```

```python
data['Height'].hist(bins=25)
plt.show()
```

### Столбчатая диаграмма

```python
data_gr_height = data.groupby('Index')['Height'].mean()
print(data_gr_height)
```

```python
data_gr_height.plot(kind='bar', title = 'Средний рост по индексу массы тела')
plt.show()
```

### Загрузка данных по качеству атмосферного воздуха

```python
aq = pd.read_csv('aq.csv')
print(aq)
```

### Линейный график

```python
aq.plot(x = 'Date', y = 'NO2', rot = 30)
plt.show()
```

### Диаграмма рассеяния

```python
data.plot(x = 'Height', y = 'Weight', kind='scatter')
```

### Добавление уровней

```python
data[data['Gender']=='Male']['Height'].hist(bins=12, alpha=0.7)
data[data['Gender']=='Female']['Height'].hist(bins=12, alpha=0.7)
plt.legend(['Male', 'Female'])
plt.show()
```
