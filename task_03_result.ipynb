```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('heart.csv')
X =df[["age","restecg","exng","oldpeak"]].values
y = df['output'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 14, stratify = y)
neighbors = np.arange(1,31)
train_acc = {}
test_acc = {}
for neighbor in neighbors:
    knn = KNeighborsClassifier(n_neighbors=neighbor)
    knn.fit(X_train, y_train)
    train_acc[neighbor] = knn.score(X_train, y_train)
    test_acc[neighbor] = knn.score(X_test, y_test)

print(neighbors)
print(train_acc)
print(test_acc)
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
heart_df = pd.read_csv("heart.csv")
Y = heart_df['output'].values
X = heart_df[['age','trtbps']].values
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=21, stratify=Y)
train_accuracies = {}
test_accuracies = {}
for neighbor in range(1, 31):  #  1  30
    knn = KNeighborsClassifier(n_neighbors=neighbor)
    knn.fit(X_train, y_train)
    train_accuracies[neighbor] = knn.score(X_train, y_train)
    test_accuracies[neighbor] = knn.score(X_test, y_test)
print(neighbors)
print(train_accuracies)
print(test_accuracies)
#    
neighbors = list(range(1, 31))
#  
plt.figure(figsize=(12, 6))
plt.title('KNN:     ')
plt.plot(neighbors, list(train_accuracies.values()), label='Train accuracies', marker='o')
plt.plot(neighbors, list(test_accuracies.values()), label='Test accuracies', marker='x')
plt.legend()
plt.grid()
plt.ylim(0, 1)
plt.show()
```
