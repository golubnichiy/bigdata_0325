### Словарь: категория - значение

```python
cat = {}
for i in post['category']:
    cat.update({i: []})
    for j in i.split('|'):
        cat[i].append(j)
print(cat)
```

```python
update_data = []
for i in cat:
    dummy = post[post["category"] == i]
    id = dummy["_id"].values[0]
    title = dummy["title"].values[0]
    post_type = dummy[" post_type"].values[0]
    for j in cat[i]:
        dict1 = {}
        dict1.update({"_id": id})
        dict1.update({"title": title})
        dict1.update({"category": j})
        dict1.update({" post_type": post_type})
        update_data.append(dict1)
    
```

```python
post1 = pd.DataFrame(update_data)
post1.head(10)
```
